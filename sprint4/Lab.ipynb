```python
import csv

import pandas as pd

import matplotlib.pyplot as plt

import seaborn as sns

collected_data = []

for i in range (5):
    name = str(input("Name: "))
    numerical = int(input("How many hours of screen time do your parents give you per day? "))
    categorical = str(input("What is your favorite music genre?"))

    collected_data.append([name, numerical, categorical])
    print(collected_data)

with open("collected_data.csv", mode="w", newline = "") as file:
    writer = csv.writer(file)
    writer.writerow(["Name", "numofhoursofscreentime", "music_genre"])
    writer.writerows(collected_data)




```

SCREEN LIMITS


```python

data = pd.read_csv("collected_data.csv")

plt.bar(data["Name"], data["numofhoursofscreentime"])
plt.title("Comparing students screen time limits.")
plt.xlabel("Name")
plt.ylabel("Hours")
plt.show()

```

MUSIC GENRE



```python

data = pd.read_csv("collected_data.csv")

sns.countplot(data=data, x="music_genre")
plt.title("Comparing music tastes")
plt.show()

```
